```python
import pandas as pd

# 构造DataFream，创建学生成绩信息
df =pd.DataFrame([
    {'姓名': '张三', '语文': 90, '数学': 80, '英语': 70},
    {'姓名': '李四', '语文': 85, '数学': 95, '英语': 85},
    {'姓名': '王五', '语文': 99, '数学': 81, '英语': 95},
    {'姓名': '赵六', '语文': 82, '数学': 90, '英语': 83},
    {'姓名': '王二', '语文': 96, '数学': 85, '英语': 75},
    {'姓名': '王七', '语文': 92, '数学': 89, '英语': 95},
])

# 统计计算
print(f"语文的最高分：{df['语文'].max()}，最低分：{df['语文'].min()}，平均分：{df['语文'].mean():.2f}")
print(f"数学的最高分：{df['数学'].max()}，最低分：{df['数学'].min()}，平均分：{df['数学'].mean():.2f}")
print(f"英语的最高分：{df['英语'].max()}，最低分：{df['英语'].min()}，平均分：{df['英语'].mean():.2f}")
```
