```python
import pandas as pd

# DataFrame的几种创建方式

# 0. 读取数据源
df0 = pd.read_csv("data/sales.csv", usecols=['产品类别', '产品名称', '单价'],  nrows=10)

# 1. 字典列表
df1 = pd.DataFrame([
    {'姓名': '张三', '语文': 90, '数学': 80, '英语': 70},
    {'姓名': '李四', '语文': 85, '数学': 95, '英语': 85},
    {'姓名': '王五', '语文': 99, '数学': 81, '英语': 95}
], index=['a', 'b', 'c'])
# 添加新列：
df1['总成绩'] = df1['语文'] + df1['数学'] + df1['英语']

# 2.字典存储列表值
df2 = pd.DataFrame({
    '姓名': ['张三', '李四', '王五'],
    '语文': [90, 85, 99],
    '数学': [80, 95, 81],
    '英语': [70, 85, 95]
})

# 3. 元组列表 or 列表列表 + 列名
df3 = pd.DataFrame([
    ('张三', 90, 80, 70),
    ('李四', 85, 95, 85),
    ('王五', 99, 81, 95)
], columns=['姓名', '语文', '数学', '英语'], index=['a', 'b', 'c']) # 自定义索引

df1
```

```python
# DataFrame的常见属性
df1.info()
df1.columns
df1.values
df1.dtypes  # 获取每列的数据类型
df1.size    # 获取单元格数量
df1.shape   # 获取行数和列数
df3.index
```

```python
# Series的几种创建方式

s = pd.Series([10, 20, 30, 40, 50]) # 列表创建
s = pd.Series((10, 20, 30, 40, 50), index=['a', 'b', 'c', 'd', 'e']) # 元组创建, 自定义索引
s = pd.Series({'a': 10, 'b': 20, 'c': 30, 'd': 40, 'e': 50}) # 字典创建, 键代表索引, 值代表数据
s = df1['语文'] # 从DataFrame中提取一列数据, 作为Series
```

```python
# Series的常见属性
s.values    # 获取Series的所有值
s.size      # 获取Series的元素数量
s.dtypes    # 获取Series的元素类型
s.index     # 获取Series的索引
s.shape     # 获取Series的行数
s.name      # 获取Series的列名 -- 前提是数据是从DataFrame中提取的一列
```
