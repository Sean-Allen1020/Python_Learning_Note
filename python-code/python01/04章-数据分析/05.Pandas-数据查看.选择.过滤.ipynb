```python
import pandas as pd

# 读取csv文件
df = pd.read_csv("data/sales.csv", usecols=["订单号", "产品类别", "产品名称", "销售数量", "单价", "支付方式"])
# 查看
df.head(10)
df.describe()   # 显示数据统计信息，如数据条数，最小值，最大值，平均值，中位数，标准差等
df.info()
# 选择列
df[["产品类别", "产品名称", "单价"]]
# 选择行
df.iloc[0:6:1]  # 根据行号筛选 不包尾
df.loc[0:6]     # 根据索引筛选 包尾
# 过滤
df[df['销售数量'] >= 10]  # 销售数量大于等于10
df[df['产品类别'].isin(['食品', '图书'])]  # 产品类别是食品或图书
df[df['单价'].between(100, 200)]  # 单价在100到200之间
df[(df['销售数量'] >= 8) & (df['单价'] >= 10)]    # 同时满足销售数量大于等于8和单价大于等于10
df[(df['支付方式'].isin(['支付宝', '微信'])) & (df['产品类别'].isin(['食品', '服装']))]
```
