```python
import pandas as pd

df = pd.read_csv("data/sales02.csv", nrows=10)

# 升序排序
df.sort_values(by="销售数量", ascending=True)  # 默认升序，ascending可以不写
# 降序排序
df.sort_values(by="销售数量", ascending=False)
# 多列排序
df.sort_values(by=["销售数量", "单价"], ascending=[True, False])
```

```python
# 数据分组
df.groupby("产品类别")["销售额"].count()  # 计算每个产品类别下的订单数量
df.groupby("产品类别")['销售额'].sum()  # 计算每个产品类别下的销售额总和
df.groupby("产品类别")["销售额"].mean()  # 计算每个产品类别下的销售额平均值
df.groupby("产品类别")["销售额"].max()  # 计算每个产品类别下的销售额最大值
df.groupby("产品类别")["销售额"].min()  # 计算每个产品类别下的销售额最小值
df.groupby("产品类别")["销售额"].agg(["mean", "max", "min"])  # 计算每个产品类别下的销售额平均值、最大值、最小值
df.groupby("产品类别").agg({"销售数量": 'sum', "销售额": 'sum', '单价': 'mean'})  # 计算每个产品类别下的销售数量总和、销售额总和、单价平均值
```
